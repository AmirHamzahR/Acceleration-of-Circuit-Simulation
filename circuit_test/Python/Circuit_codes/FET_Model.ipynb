```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vds = np.linspace(1,1000,1000) * 0.001
```

```python
beta = 1 
vt = 0.4
LAMBDA = 0.1

vgs=1

id1 = beta*(vgs-vt-vds/2)*vds*(1+LAMBDA*vds)

id2 = (beta/2) * np.power((vgs - vt),2) * (1+LAMBDA*vds)
```

```python
plt.plot(vds, id1)
plt.plot(vds, id2)
```

```python
def fet(vd, vg, vs, vb):
  vgs = vg - vs
  vds = vd - vs
  if (vgs>vt) & (vds <= vgs-vt):
    id = beta*(vgs-vt-vds/2)*vds*(1+LAMBDA*vds)
  elif (vgs>vt) & (vds > vgs-vt):
    id = (beta/2) * np.power((vgs - vt),2) * (1+LAMBDA*vds)
  else: 
    id=0
  return id
```

```python
y = np.zeros(len(vds))

for i in range(len(vds)):
  y[i] = fet(vds[i], 1.0, 0, 0)

plt.plot(vds, y)

```

```python
y = np.zeros(len(vds))

vgs = np.linspace(1,10,10)*0.1

for j in range(len(vgs)):
  for i in range(len(vds)):
    y[i] = fet(vds[i], vgs[j], 0, 0)
  plt.plot(vds, y)

```
